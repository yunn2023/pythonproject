```python
# 행정구역 컬럼을 제거하는건지 그냥 그대로 둔 채로 분리하는 건지 모르겠어서 일단 유지함..
```

```python
# 시,구,동 컬럼을 만들어 분리해 넣기
# 시
data['행정구역'].str.split(' ').str[0]
```

```python
# 시,구,동 컬럼을 만들어 분리해 넣기
# 구
data['행정구역'].str.split(' ').str[1]
```

```python
# 시,구,동 컬럼을 만들어 분리해 넣기
# 동
data['행정구역'].str.split(' ').str[2]
```

```python
# 행정구역 컬럼 오른쪽에 시,구,동 컬럼이 추가되도록 위치 지정하여 컬럼 추가하기
data.insert(1,'시',data['행정구역'].str.split(' ').str[0])
data.insert(2,'구',data['행정구역'].str.split(' ').str[1])
data.insert(3,'동',data['행정구역'].str.split(' ').str[2])
```
