```python
# 필요한 라이브러리 import 하기
import requests
from bs4 import BeautifulSoup

# 크롤링할 웹페이지의 URL 지정하기
url = 'http://www.cgv.co.kr/movies/?lt=1&ft=0'

# response.get() 메서드를 사용하여 URL의 HTML 소스 가져오기
response = requests.get(url)
html = response.text

# BeautifulSoup으로 html 소스를 python 객체로 변환하기
# BeautifulSoup 객체를 생성하여 HTML 소스를 파싱하기
soup = BeautifulSoup(html, 'html.parser')

# 각 영화 정보를 담고 있는 div 태그 선택
movie_list = soup.select('div.sect-movie-chart > ol > li')

# 각 영화 정보에서 제목, 평점, 개봉일 등 추출
for movie in movie_list:
    title = movie.select_one('div.box-contents > a > strong').text
    score = movie.select_one('div.box-contents > div > strong').text
    open_date = movie.select_one('div.box-contents > span.txt-info:nth-of-type(1)').text
    
    print(f'제목: {title}, 평점: {score}, 개봉일: {open_date}')
```

```python
# 필요한 라이브러리 import 하기
import requests
from bs4 import BeautifulSoup

# 크롤링할 웹페이지의 URL 지정하기
url = "https://music.bugs.co.kr/chart"

# response.get() 메서드를 사용하여 URL의 HTML 소스 가져오기
response = requests.get(url)
html = response.text

# BeautifulSoup으로 html 소스를 python 객체로 변환하기
# BeautifulSoup 객체를 생성하여 HTML 소스를 파싱하기
soup = BeautifulSoup(html, 'html.parser')

# 곡 제목과 아티스트명 가져오기
titles = []
artists = []

# select() 메서드를 사용하여 곡 제목과 아티스트명 추출하기
for title in soup.select('p.title'):
    titles.append(title.text.strip())

for artist in soup.select('p.artist'):
    artists.append(artist.text.strip())

# 결과 출력하기
for i in range(len(titles)):
    print(f"{i+1}위 {titles[i]} - {artists[i]}")
```
