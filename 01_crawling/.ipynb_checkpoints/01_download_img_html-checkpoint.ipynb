### 03. 검색어 넣어서 html 페이지 가져오기
#### 01. urllib 이용

```python
'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%B2%9A%EA%BD%83'
```

```python
from urllib import request, parse
```

```python
query = input('검색할 키 >>> ')
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='+query
request.urlopen(url).read()
```

```python
url = 'https://search.naver.com/search.naver?'
query = input('검색할 키 >>> ')
values = {
    'where':'nexearch', 
    'sm':'top_hty', 
    'fbm':1, 
    'ie':'utf8', 
    'query':query
}
params = parse.urlencode(values)
url = url + params
print(url)
```

```python
request.urlopen(url).read().decode('utf8')
```

#### 02. requests 이용

```python
import requests
```

```python
query = input('검색할 키 >>>')
url = 'https://search.naver.com/search.naver?'
values = {
    # 'where':'nexearch', 
    # 'sm':'top_hty', 
    # 'fbm':1, 
    # 'ie':'utf8', 
    'query':query
}

r = requests.get(url, params=values)
r.text
```
