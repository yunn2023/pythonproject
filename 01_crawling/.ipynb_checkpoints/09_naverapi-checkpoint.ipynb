```python
# 네이버 검색 API 예제 - 블로그 검색

# import os # import sys # 안 쓰는 코드 주석 처리
import urllib.request
import json  # 필요한 코드 추가함

client_id = "EMUYyDHGQ01TnD838z8B"
client_secret = "suFf0hvkBl"

query = input('검색할 단어 >>')
encText = urllib.parse.quote(query)
print(encText)

# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# display = 100 으로 변경해보기
url = "https://openapi.naver.com/v1/search/blog?display=100&query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
data = response_body.decode('utf-8')
print(type(data))
```

```python
# 파일에 저장되어 있는 데이터를 불러올 때는 load
# 변수에 저장되어 있는 데이터를 불러올 때는 loads
result = json.loads(data)
print(type(result))
```
