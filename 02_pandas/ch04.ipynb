```python
#     :  
def recode_sex(sex): #  
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex']
```

```python
# tips[].apply()
tips['sex'].apply(recode_sex)
```

```python
tips['color'] = tips['sex'].apply(recode_sex)
```
