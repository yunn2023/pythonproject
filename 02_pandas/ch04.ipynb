```python
# 다변량 데이터로 다변량 그래프 그리기: 산점도 그래프
def recode_sex(sex): # 함수 선언
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex']
```

```python
# tips[함수인자].apply(함수이름)
tips['sex'].apply(recode_sex)
```

```python
tips['color'] = tips['sex'].apply(recode_sex)
```

```python
tips
```

```python
# 다변량 데이터로 다변량 그래프 그리기: 산점도 그래프
scatter_plot = plt.figure()
ax1 = scatter_plot.add_subplot(111)
ax1.scatter(tips['total_bill'], 
            tips['tip'], 
            s = tips['size']*10, 
            c = tips['color'], 
            alpha = 0.5)
# 범례 추가하려면 legend 사용
# ax1.set_title('산점도') #한글 깨짐
ax1.set_title('Scatter Plot') 
```
