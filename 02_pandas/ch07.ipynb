```python
weather = pd.read_csv('data/weather.csv')
weather
```

```python
weather.head(4)
```

```python
weather.columns
```

```python
# 가상 데이터의 여러 열을 하나로 정리하기 - melt, pivot_table 메서드
```

```python
# melt
weather_melt = pd.melt(weather, id_vars=['id','year','month','element'], 
                       var_name='day', value_name='temp')
```

```python
weather_melt.head(3)
```

```python
# pivot_table 메서드
weather_melt.pivot_table(index=['id','year','month','day'], 
                         columns=['element'], 
                         values=['temp'])
```

```python
# pivot_table 메서드
# 대괄호를 빼면 한 단계가 내려간다..?
weather_melt.pivot_table(index=['id','year','month','day'], 
                         columns=['element'], 
                         values='temp')
```

```python
# pivot_table 메서드
# 대괄호를 빼면 한 단계가 내려간다..?
weather_melt.pivot_table(index=['id','year','month','day'], 
                         columns='element', 
                         values='temp')
```

```python
weather_pivot = weather_melt.pivot_table(index=['id','year','month','day'], 
                                         columns='element', 
                                         values='temp')
weather_pivot
```

```python
weather_pivot.head(3)
```

```python
# 중복 데이터 처리하기
```

```python
billboard = pd.read_csv('data/billboard.csv')
billboard
```

```python
billboard.head(1)
```

```python
billboard.columns
```

```python
billboard_melt = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], 
                         var_name='week', value_name='rating')
billboard_melt
```

```python
billboard.shape
```

```python
billboard_melt.shape
```

```python
billboard.shape, billboard_melt.shape
```

```python
songs = billboard_melt[['year','artist','track','time']]
songs
```

```python
songs.shape
```

```python
# drop_duplicates(): 데이터프레임의 중복 데이터 제거하기
songs.drop_duplicates(inplace=True)
```

```python
songs.shape
```

```python
songs['id'] = range(len(songs))
```

```python
songs
```

```python
songs.columns
```

```python
billboard_rating = billboard_melt.merge(songs, on=['year','artist','track','time'])
billboard_rating
```

```python
billboard_rating.drop(columns=['year','artist','track','time'], inplace = True)
billboard_rating.drop
```
