```python
ebola = pd.read_csv('data/country_timeseries.csv')
```

```python
ebola.info()
```

```python
ebola.isna()
# na값이면 True, na값이 아니면 False
```

```python
ebola.isna().sum()
```

```python
# 누락값 처리하기
```

```python
# 누락값 처리하기 - 변경, 삭제
ebola.fillna(0)
# 누락값을 0으로 채운다.
```

```python
ebola.fillna(0).head(2)
```

```python
ebola['Cases_Liberia'].fillna(0)
```

```python
ebola.head(5)
```

```python
ebola.fillna(method='ffill')
# 누락값이 나타나기 전의 값으로 누락값이 변경됨.
# 단점: 초반(ex. 0, 1행) 값이 누락값인 경우에는 처리하지 못함.
```

```python
ebola.fillna(method='bfill')
# 누락값이 나타난 이후의 첫 번째 값으로 앞쪽의 누락값이 모두 변경됨.
# 단점: 마지막 값이 누락값인 경우에는 처리하지 못함.
```

```python
ebola.interpolate()
# 누락값 양쪽에 있는 값을 이용하여 중간값을 구한 다음 누락값을 처리한다.
```

```python
ebola.fillna(ebola.mean())
# 데이터에 숫자만 있는 게 아니라서 mean() 실행 시 에러 발생 가능.
```

```python
ebola['Cases_Guinea'].fillna(ebola['Cases_Guinea'].mean())
```

```python
# 누락값 삭제하기
# dropna 메서드 사용
# 누락값이 포함된 행들이 모두 삭제되기 때문에 분석자가 잘 판단하여 삭제해야 함.
```

```python
# dropna 메서드 사용
ebola.dropna()
# 누락값이 포함된 행들이 모두 삭제되기 때문에 많은 데이터가 삭제되었다.
```

```python
ebola.shape()
```

```python
ebola.dropna(axis=0)
```

```python
ebola.dropna(axis=1)
```

```python
# 누락값이 포함된 데이터 계산하기
```

```python
ebola['Cases_Guinea'] + ebola['Cases_Liberia']
```

```python
# sum 메서드 사용할 때 누락값을 무시한 채 계산하려면
# skipna 인자값을 True로 설정하기
print(ebola['Cases_Guinea'].sum(skipna = True))
```

```python
print(ebola['Cases_Guinea'].sum(skipna = False))
```
