```python
# 한글 깨짐 해결하기
# 1. 폰트 파일로 FontProperties()
# D2Coding.ttf 파일을 복사하여 02_pandas 폴더에 붙여넣기
import matplotlib.font_manager as fm

path = 'D2Coding.ttf'
fontprop = fm.FontProperties(fname=path, size=14)

data = np.random.randint(-100, 100, 50).cumsum()
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격', fontproperties=fontprop)
plt.xlabel('시간(분)', fontproperties=fontprop)
```

```python
# 한글 깨짐 해결하기
# 2. rcParams[] 를 설정
import matplotlib

print('설정 파일 위치:', matplotlib.matplotlib_fname())
print('기본 설정 폰트:', plt.rcParams['font.family'])
print('기본 폰트 크기:', plt.rcParams['font.size'])

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 12
print('기본 설정 폰트:', plt.rcParams['font.family'])
print('기본 폰트 크기:', plt.rcParams['font.size'])
```

```python
# 한글 깨짐 해결하기
# 2. rcParams[] 를 설정
import matplotlib

print('설정 파일 위치:', matplotlib.matplotlib_fname())
print('기본 설정 폰트:', plt.rcParams['font.family'])
print('기본 폰트 크기:', plt.rcParams['font.size'])

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False # -가 깨지지 않게 하기 위해 코드 추가함

print('기본 설정 폰트:', plt.rcParams['font.family'])
print('기본 폰트 크기:', plt.rcParams['font.size'])

data = np.random.randint(-100, 100, 50).cumsum()
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```
