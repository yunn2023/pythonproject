```python
# 1. 데이터 집계
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/gapminder.tsv', sep='\t')
df
```

```python
# 분할-반영-결합 과정 살펴보기 - groupby 메서드
```

```python
df.groupby('year').lifeExp.mean()
```

```python
df.groupby('year')['lifeExp'].mean()
# 같은 결과
```

```python
df.groupby('year')['lifeExp'].mean().to_frame()
```

```python
df.groupby('year')['lifeExp'].mean().to_frame().reset_index()
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum/n
```

```python
df.groupby('year')['lifeExp'].agg(my_mean)
```

```python
import numpy as np
```

```python
# 여러 개의 집계 메서드 한 번에 사용하기
# 집계 메서드를 리스트, 딕셔너리에 담아 전달하기
```

```python
# 집계 메서드를 리스트에 담아 전달하기
df.groupby('year')['lifeExp'].agg([np.count_nonzero, np.mean, np.std])
```

```python
# 집계 메서드를 딕셔너리에 담아 전달하기
df.groupby('year').agg({'lifeExp':'mean',
                        'pop':'median',
                        'gdpPercap':'median'})
```

```python
# 집계 메서드를 딕셔너리에 담아 전달하기
df.groupby('year').agg({'lifeExp':'median',
                        'pop':'median',
                        'gdpPercap':'median'})
```
