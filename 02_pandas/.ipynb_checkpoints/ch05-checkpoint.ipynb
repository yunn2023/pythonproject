```python
# 분석하기 좋은 데이터란?
# [깔끔한 데이터(Tidly Data)의 조건]

# 데이터 분석 목적에 맞는 데이터를 모아 새로운 표(Table)를 만들어야 한다.
# 측정한 값은 행(row)을 구성해야 한다.
# 변수는 열(column)을 구성해야 한다.
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/concat_1.csv')
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# concat 메서드로 데이터 연결하기
pd.concat([df1,df2,df3])
```

```python
# concat 메서드로 데이터 연결하기
pd.concat([df1,df2,df3], axis=1)
```

```python
new = pd.Series(['n1','n2','n3','n4'])
new
```

```python
pd.concat([df1, new])
# pd.concat([df1, new], axis=0) # 같은 결과: default가 axis=0
```

```python
pd.concat([df1, new], axis=1) 
# 0,1,2,3을 기준으로 데이터를 붙여서 문제 없이 생성됨.
```

```python
new_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A','B','C','D'])
new_df
```

```python
pd.concat([df1, new_df])
```

```python
pd.concat([df1, new_df], ignore_index=True)
```

```python
pd.concat([df1,df2,df3], axis=1, ignore_index=True)
```

```python
pd.concat([df1,df2,df3], axis=0, ignore_index=True)
```

```python
df1.columns
```

```python
df1.columns = ['A','B','C','D']
df2.columns = ['E','F','G','H']
df3.columns = ['A','C', 'F', 'H']
```

```python
pd.concat([df1,df2,df3], axis=0, join='outer')  # 외부 조인(outer join)
```

```python
pd.concat([df1,df2,df3], axis=0, join='inner')  # 내부 조인(inner join)
```

```python
pd.concat([df1,df3], axis=0, join='inner')  # 내부 조인(inner join)
```

```python
# merge 메서드 사용하기
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited = pd.read_csv('data/survey_visited.csv')
```

```python
person.head(2)
```

```python
site.head(2)
```

```python
survey.head(2)
```

```python
visited.head(2)
```

```python
visited
```

```python
v_subset = visited.loc[[0,2,6]]
v_subset
```

```python
# merge 메서드 사용하기
# pd.merge()
site.merge(v_subset, left_on='name', right_on='site')
```

```python
site.merge(visited, left_on='name', right_on='site')
```

```python
person.merge(survey, left_on='ident', right_on='person')
```

```python
visited.merge(survey, left_on='ident', right_on='taken')
```

```python
ps = person.merge(survey, left_on='ident', right_on='person')
```

```python
vs = visited.merge(survey, left_on='ident', right_on='taken')
```

```python
ps.head(1)
```

```python
vs.head(1)
```

```python
ps.merge(vs, 
         left_on=['ident','taken','quant','reading'], 
         right_on=['person','ident','quant','reading'])
```

```python
ps.merge(vs, 
         left_on=['ident','taken','quant','reading'], 
         right_on=['person','ident','quant','reading'], how='outer')
```
