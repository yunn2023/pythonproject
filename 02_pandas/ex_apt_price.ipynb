```python
# melt를 이용하여 월별데이터를 '기간' 컬럼으로, 가격은 '평당분양가격'으로 처리하여 first_df_melt로 저장
first_df_melt = first_df.melt(id_vars='지역', var_name='기간', value_name='평당분양가격')
first_df_melt.head(5)
```

```python
# 기간 컬럼을 연도, 월 컬럼으로 분리
first_df_melt['기간'].str.split('년')
```

```python
first_df_melt['기간'].str.split('년').str[0]
```

```python
first_df_melt['기간'].str.split('년').str[1]
```

```python
# 기간 컬럼을 연도, 월 컬럼으로 분리
```

```python
first_df_melt['연도'] = first_df_melt['기간'].str.split('년').str[0].astype(int) # int로 변경
```

```python
first_df_melt['월'] = first_df_melt['기간'].str.split('년').str[1].astype(int) # int로 변경
```

```python
# last_df 전용 면적에서 전체 인 데이터만 last_df_prepare로 저장
last_df_prepare = last_df[last_df['전용면적']=='전체'].copy()
last_df_prepare.head(3)
```

```python
# 지역명 -> 지역 컬럼으로 변경하기
last_df_prepare.columns = ['지역명','연도','월','분양가격','평당분양가격','전용면적']
last_df_prepare.columns
```

```python
# 지역명 -> 지역 컬럼으로 변경하기
last_df_prepare.rename(columns={'지역명':'지역'}, inplace=True)
last_df_prepare.columns
```

```python
first_df_melt.rename(columns={'지역명':'지역'}, inplace=True)
first_df_melt.columns
```

```python
# concat으로 데이터 합치기
first = first_df_melt[['지역','연도','월','평당분양가격']]
last = last_df_prepare[['지역','연도','월','평당분양가격']]
df = pd.concat([first,last])
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```
