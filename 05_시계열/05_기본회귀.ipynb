```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
bostondf.info()
```

```python
bostondf.head(3)
```

```python
bostondf.columns
```

```python
# 2개의 행과 4개의 열을 가진 subplots를 이용, axs는 4x2개의 ax를 가짐.
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)
lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
for i,feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    # 시본의 regplot을 이용해 산점도와 선형 회귀 직선을 함께 표현
    sns.regplot(x=feature, y='MEDV', data=bostondf, ax=axs[row][col])
```

```python
y = bostondf['MEDV']
x = bostondf.drop(columns=['MEDV'])
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=156)
```

```python
x_train, x_test, y_train, y_test
```

```python
# 선형 회귀 OLS로 학습/예측/평가 수행.
lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)
print(f'mse:{mse}, rmse:{rmse}, r2:{r2}')
```

```python
print(f'절편(바이어스): {lr.intercept_}, \n회귀계수(가중치): {np.round(lr.coef_, 1)}')
```

```python
# 회귀 계수를 큰 값 순으로 정렬하기 위해 Series로 생성, 인덱스 칼럼명에 유의
coeff = pd.Series(data=np.round(lr.coef_, 1), index=x.columns)
coeff
```

```python
coeff.sort_values(ascending=False)
```

```python
from sklearn.model_selection import cross_val_score # 교차검증
```

```python
lr = LinearRegression()
y = bostondf['MEDV']
x = bostondf.drop(columns=['MEDV'])

# cross_val_score()로 5 폴드 세트로 MSE를 구한 뒤 이를 기반으로 다시 RMSE 구함.
neg_mse = cross_val_score(lr,x,y,scoring='neg_mean_squared_error', cv=5)
neg_mse
# 로 반환된 값은 모두 음수.
```

```python
-neg_mse
```

```python
mse = -neg_mse
rmse = np.sqrt(mse)
np.mean(rmse)
```
