```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(mnist.data, mnist.target, test_size=0.1)
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
accuracy_score(y_test,pred)
```

### pickle 사용하기 & 모델로 학습시키기

```python
# pickle: 객체의 형태를 그대로 유지하면서 
#         파일에 저장하고 불러올 수 있게 하는 모듈
```

```python
import pickle
```

```python
# mnist.model을 생성하여 학습시키기
```

```python
f = open('mnist.model', 'wb')
pickle.dump(clf,f)
f.close()
```

```python
f = open('mnist.model', 'rb')
model = pickle.load(f)
f.close()
```

```python
for path in glob.glob('./mnist/*.png'):
    # print(path)
    img = Image.open(path).convert('L')
    # print(img)
    plt.imshow(img, cmap='gray')
    img_result = np.resize(img,(1,784))
    # print(img_result)
    img_result = 255 - img_result
    # print(img_result)
    # img_result = np.resize(img_result,(28,28))
    # plt.imshow(img_result, cmap='gray')
    pred = model.predict(img_result)
    print(pred)
    plt.show()
```
